```python
#Bai 5
import math
a = float(input())
b = float(input())
c = float(input())
if a == 0 :
    if b == 0:
        if c == 0:
            print ("Phuong trinh vo so nghiem")
        else:
            print ("Phuong trinh vo nghiem")
    else:
        x = -c/b
        print ("Nghiem cua phuong trinh la",x)
else:
    delta = b*b - 4*a*c
    if delta > 0:
        x1 = (-b + math.sqrt(delta))/(2*a)
        x2 = (-b - math.sqrt(delta))/(2*a)
        print ("Phuong trinh co 2 nghiem phan biet")
        print ("Nghiem cua phuong trinh la",x1)
        print ("Nghiem cua phuong trinh la",x2)
    elif delta == 0:
        x = -b / (2*a)
        print ("Phuong trinh co nghiem kep")
        print ("Nghiem cua phuong trinh la",x)
    else :
        real = -b / (2*a)
        imag = math.sqrt(-delta) / (2*a)
        print ("Phuong trinh co nghiem phuc")
        print (f"x1 = {real} + {imag}i")
        print (f"x2 = {real} - {imag}i")

```
