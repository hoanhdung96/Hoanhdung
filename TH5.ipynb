```python
#W5A1
def maxoftwo(a,b):
    if a > b :
        return a
    elif b > a  :
        return b
print ( maxoftwo(18 , 36))
```

```python
#W5A3
def ngto(a):
    if a < 2 :
        return False
    for i in range ( 2 , int(a**0.5) + 1 ):
        if a % i == 0 :
            return False
    return True
print ( ngto(29))
```

```python
#W5A4
def so_hoan_hao(n):
    if n <= 1 :
        return False
    tong = 0
    for i in range ( 1 , n // 2 + 1 ): #không có ước nào lớn hơn n // 2 ngoại trừ n
        if n % i == 0 :
            tong += i
    return tong == n
print ( so_hoan_hao(28))

```

```python
#W5A6
def giaithua(n):
    gt = 1
    for i in range ( 1 , n + 1 ):
        gt *= i
    return gt
print ( giaithua(13))

```

```python
#W5A9
def tong_chu_so(n):
    tong = 0
    while n != 0:
        digit = n % 10
        tong += digit
        n = n // 10
    return tong

print ( tong_chu_so(1234567))
```

```python
#W5A6
def find_index(nums , k ):
    if k in nums:
        return '1'
    else :
        return '-1'
k = int(input("Nhap so k "))
nums = list(map(int, input().split()))
print (find_index(nums, k))
```

```python
#W5A7
def tinh_toan(num1 , num2):
    hieu = num1 - num2
    tong = num1 + num2
    tich = num1 + num2
    thuong = round(num1 / num2 , 2)
    return hieu, tong, tich, thuong
num1 = float(input())
num2 = float(input())
print ( tinh_toan(num1 , num2))
```
