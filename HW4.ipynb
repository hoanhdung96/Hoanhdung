```python
#W4A6
a = int(input())
b = int(input())
tong = 0
for x in range ( a , b + 1 ):
    if x >= 2:
        ngto = True
        for i in range ( 2 , int(x**0.5) + 1 ):
            if x % i == 0:
                ngto = False
                break
        if ngto :
            tong += x

print (tong)







```
