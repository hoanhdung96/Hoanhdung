```python
#W6A9
dict1 = {"a" : 1 , "b" : 2 , "c" : 3 }
dict2 = {"b" : 3 , "c" : 4 , "d" : 5 }
result = dict1.copy()
for k,v in dict2.items():
    if k in result:
        result[k] += v
    else:
        result[k] = v
print (result)
```

```python
#W6A10
nums = list(map(int, input().split()))
k = int(input())
result = []
n = len(nums)
for i in range ( n ):
    for j in range ( i + 1 , n ):
        if nums[i] + nums[j] == k :
            result.append((nums[i] , nums [j]))
print (result)


```

```python
#W6A11
t = ( 1,2,3,4,5,6,7,8,9)
even = set()
odd = set()
for i in t:
    if i % 2 == 0 :
        even.add(i)
for j in t:
    if j % 2 == 1 :
        odd.add(j)
even_tuple = tuple(sorted(even))
odd_tuple = tuple(sorted(odd))
print (even_tuple)
print (odd_tuple)
```

```python
#W6A12
nums = list(map(int, input().split()))
d = {}
for x in nums:
    if x not in d:
        d[x] = 1
    else :
        d[x] += 1
max_d = max(d.values())
num = (num for num, f in d.items() if f == max_d)
max_num = max(num)
print(max_num)



```
