```python
#W9A1
def merge_sorted_list(list1, list2):
    i = 0
    j = 0
    merged = []
    while i < len(list1) and j < len(list2):
        if list1[i] <= list2[j]:
            merged.append(list1[i])
            i += 1
        else:
            merged.append(list2[j])
            j += 1
    while i < len(list1):
        merged.append(list1[i])
        i += 1
    while j < len(list2):
        merged.append(list2[j])
        j += 1
    return merged
list1 = list(map(int, input().split()))
list2 = list(map(int, input().split()))
print(list1)
print(list2)
print(merge_sorted_list(list1, list2))








```

```python
#W9A3
def smallest_by_swap(n):
    digits = list(str(n))
    length = len(digits)
    min_right = [None] * length
    min_digit = digits[-1]
    min_idx = length - 1
    for i in range(length - 2 , -1 , -1 ):
        if digits[i+1] < min_digit:
            min_digit = digits[i+1]
            min_idx = i + 1
        min_right[i] = ( min_digit , min_idx)
    for i in range(length):
        if min_right[i] and min_right[i][0] < digits[i]:
            j = min_right[i][1]
            digits[i] , digits[j] = digits[j], digits[i]
            break
    return int(''.join(digits))
n = int(input())
print(n)
print(smallest_by_swap(n))
```

```python
#W9A4
def can_transform(a,b):
    i , j = 0 , 0
    while i < len(a) and j < len(b):
        if a[i].upper() == b[j]:
            j += 1
        i += 1
    return j == len(b)
Q = int(input("Q"))
print(Q)
for _ in range(Q):
    a = input("a").strip()
    b = input("b").strip()
    print(a)
    print(b)
    print("YES" if can_transform(a,b) else "NO")
```

```python
#W9A5
def if_list_in_list(list1, list2):
    n = len(list1)
    m = len(list2)
    if m <= n :
        return "No"
    for i in range(m - n + 1 ):
        if list2[i:i+n] == list1:
            return "Yes"
        else:
            return "No"
list1 = list(map(int, input().split()))
list2 = list(map(int, input().split()))
print(list1)
print(list2)
print(if_list_in_list(list1, list2))

```

```python
#W9A6
def next_permutation(a):
    n = len(a)
    i = n - 2
    while i >= 0 and a[i] > a[i+1]:
        i -= 1
    if i >= 0 :
        j = n - 1
        while a[j] < a[i]:
            j -= 1
        a[i], a[j] = a[j], a[i]
    a[i+1:] = reversed(a[i+1:])
    return a
n = int(input().strip())
a = list(map(int, input().split()))
res = next_permutation(a)
print(" ".join(map(str, res)))
```

```python
#W9A7
def canPlaceFlowers(flowerbed , k):
    n = len(flowerbed)
    for i in range(n):
        if flowerbed[i] == 0:
            empty_left = (i == 0) or (flowerbed[i-1] == 0)
            empty_right = (i == n -1 ) or (flowerbed[i+1] == 0)
            if empty_left and empty_right:
                flowerbed[i] == 1
                k -= 1
                if k == 0 :
                    return True
    return k <= 0
k = int(input("k"))
flowerbed = list(map(int, input("flowerbed").split()))
print(k)
print(flowerbed)
print(canPlaceFlowers(flowerbed, k))
```

```python
#W9A8
def is_path_crossing(moves):
    x , y = 0 , 0
    visited = {(0 , 0)}
    for move in moves:
        if move == "R":
            x += 1
        elif move == "L":
            x -= 1
        elif move == "U":
            y += 1
        elif move == "D":
            y -= 1
        if (x,y) in visited:
            return True
        visited.add((x,y))
    return False
moves = input().split()
print(moves)
print(is_path_crossing(moves))
```
