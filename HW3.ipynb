```python
#Bai 4
h = int(input())
for i in range(1, h + 1):
    if i == 1 :
        print ( " " * ( h - i ) + "*")
    elif i == h :
        print ( "*" * ( 2*h - 1))
    else:
        print ( " " * (h - i) + "*" + " " * (2*i - 3) + "*")

```

```python
#Bai 15
n = int(input())
x = n
cnt = 0
n = abs(n)
if n == 0 :
    cnt == 1
else:
    while n != 0 :
        cnt += 1
        n =  n // 10

print ("So chu so cua so",x,"la",cnt )
```

```python
#Bai 16
n = int(input())
temp = abs(n)
even = 0
odd = 0
if temp == 0:
    print ("0")
else :
    while temp != 0:
        digit = temp % 10
        if digit % 2 == 0:
            even += 1
        else :
            odd += 1
        temp //= 10
print ("So da nhap la",n)
print ("So chu so chan la",even )
print ("So chu so le la", odd )
```
