```python
#W6A16
n, m = list(map(int, input().split()))
matrix = []
for i in range(n):
    row = list(map(int, input().split()))
    matrix.append(row)
for i in range(n):
    for j in range(m):
        print(f"{matrix[i][j]:>4}", end="")
    print()
```

```python
#W6A19
def max_nums(nums):
    max = nums[0]
    for i in range(len(nums)):
        if nums[i] > max:
            max = nums[i]
    return max
nums = list(map(int, input().split()))
print(max_nums(nums))

```

```python
#W6A20
def positions(nums , k ):
    for i in range(len(nums)):
        if nums[i] == k:
            return i
    if k not in nums:
        return -1
nums = list(map(int, input().split()))
k = int(input())
print(positions(nums, k))



```

```python
#W6A21
text = input().split()
d = {}

for idx, word in enumerate(text):
    if word not in d:
        d[word] = []
    d[word].append(idx)
for key in d:
    d[key] = tuple(sorted(d[key]))
print(d)
```

```python
#W6A22
def max_position(nums):
    max = nums[0]
    for i in range(len(nums)):
        if nums[i] > max :
            max = nums[i]
    return i
nums = list(map(int, input().split()))
print(max_position(nums))
```

```python
#W6A23
def remove(nums , x):
    small = []
    big = []
    for i in range(len(nums)):
        if nums[i] < x :
            small.append(nums[i])
        else:
            big.append(nums[i])
    return big

nums = list(map(int, input().split()))
x = int(input())
print(remove(nums, x))
```

```python
#W6A24
def push(nums):
    zero = [x for x in nums if x == 0]
    nonzero = [x for x in nums if x != 0]
    return zero + nonzero
nums = list(map(int, input().split()))
print(push(nums))

```

```python
#W6A28
day = list(map(int, input().split()))
increase = True
decrease = True
for i in range(len(day)):
    for j in range(i + 1 , len(day)):
        if day[i] < day[j]:
            decrease = False
        elif day[i] > day[j]:
            increase = False
        else:
            increase = False
            decrease = False
if increase:
    print("Day so tang that su")
elif decrease:
    print("Day so giam that su")
else:
    print("Vo danh")


```
