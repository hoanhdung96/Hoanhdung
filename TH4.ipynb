```python
#W4A6
def is_prime(n):
    if n < 2 :
        return False
    for i in range(2, int(n**0.5) + 1 ):
        if n % i  == 0:
            return False
    return True
while True:
    try:
        line = input()
        if not line:
            break
        a,b = map(int , line.split())
        total = 0
        for num in range(a, b + 1 ):
            if is_prime(num):
                total += num
            print (total)
    except EOFError:
        break
```

```python
#W4A7
n = int(input())
i = 2
largest_prime = 0
num = n
while i*i <= num:
    while num % i == 0:
        largest_prime = i
        num //= i
    i += 1
if num > 1:
    largest_prime = num
print ("Uoc so nguyen to lon nhat cua ", n, "la", largest_prime)
```
